```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
df = pd.read_csv("OnlineNewsPopularity.csv")
```

```python
def scrap(url):
    soup = BeautifulSoup(requests.get(url).text,'html.parser')
    
    try : 
        titre = soup.h1.text
    except : 
        titre = None
    
    try : 
        content_total = soup.find(class_='article-content')
    except : 
        content_total = None
    
    
    if content_total != None : 
    
        len_img = len(re.findall("img",str(content_total)))
        content = ''

        for i in content_total.findAll('p'):
            content += i.get_text(strip=True).replace("\n", " ").replace("\'","")
    else : 
        len_img = None 
        content = None
    
    return titre, len_img, content
```

```python
t,n,c = scrap("https://mashable.com/article/easy-pranks/")
```
