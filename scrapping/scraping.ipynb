```python
def scrap(url):
    soup = BeautifulSoup(requests.get(url).text,'html.parser')
    
    try : 
        titre = soup.h1.text
    except : 
        titre = None
    
    try : 
        content_total = soup.find(class_='article-content')
    except : 
        content_total = None
    
    
    if content_total != None : 
        
        see_also = content_total.find(class_='see-also')
        content_total = BeautifulSoup(str(content_total).replace('<em>','').replace('</em>','').replace(str(see_also),' '), 'html.parser')
        content = ''

        for i in content_total.findAll('p')[:-1]:
            content += " " + i.get_text(strip=True).replace("\'","")
    else : 
        content = None

    
    return titre, content
```

```python
def applying(row):
    titre, content = scrap(row["url"])
    print("Numéro :",row.name)
    if int(row.name) % 5000 == 0:
        sleep(10)
    return pd.Series([titre, content])

df[['titre','content']] = df.apply(applying ,axis=1)
```
